```python
kaoyan = df2[df2["就业意向（考研or就业）"] == "考研"]
```

```python
kaoyan["就业单位（已签约的同学填）"]
```

```python
kaoyan["就业单位（已签约的同学填）"].dropna()
```

```python
ky = kaoyan["就业单位（已签约的同学填）"].dropna()
```

```python
n = []
for i in ky:
    i = i.split(" ")[0]
    i = i.split("（")[0]
    i = i.split("(")[0]
    i = i.split("2")[0]
    i = i.split("3")[0]
    i = i.split("1")[0]
    i = i.split("4")[0]
    i = i.split("考")[0]
    i = i.split("保")[0]
    if i.endswith("司") or i.endswith("厂") or i.endswith("局"):
        pass
    elif i == "深大":
        i = "深圳大学"
    else:
        n.append(i)
sch = pd.Series(n)
sch.replace(" ", "NaN")
sch.replace(u"无", "NaN")
sch.dropna()
print(sch)
```

```python
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['KaiTi']
mpl.rcParams['font.serif'] = ['KaiTi']
# mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题,或者转换负号为字符串
sns.set_style("darkgrid",{"font.sans-serif":['KaiTi', 'Arial']})

f = plt.figure(figsize=[15, 7], dpi=80)

lbs = sch.value_counts().index
explodes = [0.1 if i == r"中南大学" else 0 for i in lbs]
# plt.pie(df["department"].value_counts(normalize=True), labels=lbs, autopct="%1.1f%%")
# plt.pie(df["department"].value_counts(normalize=True), explode=explodes, labels=lbs, autopct="%1.1f%%", colors=sns.color_palette("Reds"))
plt.pie(sch.value_counts(normalize=True, ), explode=explodes, labels=lbs, autopct="%1.1f%%")
plt.show()
```

```python
# 解决matplotlib画图中文显示乱码问题
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['KaiTi']
mpl.rcParams['font.serif'] = ['KaiTi']
# mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题,或者转换负号为字符串
sns.set_style("darkgrid",{"font.sans-serif":['KaiTi', 'Arial']})

f = plt.figure(figsize=[25, 7], dpi=160)
plt.title("机械工程学院2020届毕业生考研学校分布")
plt.xlabel("学校")
plt.ylabel("人数")
plt.xticks(np.arange(len(sch.value_counts())), sch.value_counts().index, rotation= 'vertical')
plt.bar(np.arange(len(sch.value_counts())), sch.value_counts())
for x, y in zip(np.arange(len(sch.value_counts())) - 0.5, sch.value_counts()):
    plt.text(x, y, y)
plt.show()
```
